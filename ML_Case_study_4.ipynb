```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
for column in df.columns:
    df[column]=label.fit_transform(df[column])
df
```

```python
df.columns
```

```python
features_df=df.drop(['Churn'],axis=1)
target_df=df['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features_df,target_df,test_size=0.3,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
acc_train=[]
acc_test=[]
for k in range(1,21):
  knn=KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train,y_train)
  acc_train.append(knn.score(X_train,y_train))
  acc_test.append(knn.score(X_test,y_test))
plt.figure(figsize=(15,10))
plt.xticks(range(1,21,1))
plt.plot(range(1,21),acc_train,color='blue',label='accuracy_train_neighbors')
plt.plot(range(1,21),acc_test,color='red',label='accuracy_test_neighbors')
plt.grid()
plt.title('accuracy score for diff value of k')
plt.show()
```

```python
#Train kNN classifier model again for 'k = 15' with under samled train set.
knn_1=KNeighborsClassifier(n_neighbors=1)
knn_1.fit(X_train,y_train)

#Call the 'Score()' function to check the accuracy score of the train and test set.
print('Train set accuracy:',knn_1.score(X_train,y_train))
print('Test set accuracy:',knn_1.score(X_test,y_test))
```

```python
#Display the presision, recall and f1-score values.
from sklearn.metrics import classification_report
print(classification_report(y_test, knn_1.predict(X_test)))
```

```python
#Perform random undersampling on train set 
from imblearn.over_sampling import RandomOverSampler

ros= RandomOverSampler(random_state=42)
X_ros_train, y_ros_train = ros.fit_resample(X_train,y_train)
```

```python
#Check the type and shape of 'X_ros_train' and 'y_ros_train' datasets.
print(type(X_ros_train), X_ros_train.shape)
print(type(y_ros_train), y_ros_train.shape)
```

```python
#Find the number of occurences of class '0' and class '1' values in the 'y_ros_train' NumPy array.
y_ros_train.value_counts()
```

```python
# Train kNN classifier model for 'k = 15'
from sklearn.neighbors import KNeighborsClassifier
knn_ros=KNeighborsClassifier(n_neighbors=15)
knn_ros.fit(X_ros_train,y_ros_train)
```

```python
#call the 'score()' function to check the accuracy score of the train and test set.
print('Train set accuracy:', knn_ros.score(X_ros_train, y_ros_train))
print('Test set accuracy:', knn_ros.score(X_test, y_test))
```

```python
#Display the precission, recall and f1-score values.
print(classification_report(y_test, knn_ros.predict(X_test)))
```
